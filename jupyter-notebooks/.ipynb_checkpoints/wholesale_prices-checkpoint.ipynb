In this report, we will construct one or more plots showing how the relationships between the type of apple, the average wholesale price of that type, and the England and Wales orchard space for each type has changed over the period 2004-2015.

#### 1. Import the two datasets

```python
# import pandas and numpy
import pandas as pd
import numpy as np
import matplotlib as plt

%matplotlib inline

# note: deactivated chained assignment warning
pd.options.mode.chained_assignment = None
```

```python
# read fruitveg csv into dataframe with separator as tab, no header, and removed error for lines
df_fruitveg = pd.read_csv('data/fruitveg.csv', sep='\t', header=0, error_bad_lines=False)

# check
df_fruitveg.head()
```

```python
# read orchfruit csv into dataframe with separator as tab, no header, and removed error for lines
df_orchfruit = pd.read_csv('data/orchfruit_ap&pr_30may13.csv', sep=',', header=0, error_bad_lines=False)

# check
df_orchfruit.head()
```

#### 2. Identify and handle missing data

```python
# note: there are multiple rows with missing values in the fruitveg dataset, decided to 
# keep missing values for average prices (which might be needed in analysis), but forward fill 
# the values for types, and remove the Quality and Units columns, noting that there are multiple
# qualities per apple variety

# drop some columns for clarity
df_fruitveg.drop(['Quality', 'Units'], axis=1, inplace=True)

# rename some columns for clarity, note that 2004 is first year
df_fruitveg.rename(columns={'2004':'Type', 'Unnamed: 1':'Apple variety'}, inplace=True)

# forward fill empty values in Type column
df_fruitveg['Type'].ffill(inplace=True)

# check
df_fruitveg.head()
```

```python
# note: there is no distinct column for year data in the fruitveg dataset, so decided to add 
# one and get the corresponding values from the Type column using a function

# add column to hold year data, fill it with NaN
df_fruitveg['Year'] = np.nan

# replace first value with first year (as noted when renaming columns), for function to work properly
df_fruitveg.loc[0,('Type')] = '2004'

# define function to find and copy year values in each row of the Type column, this could also 
# be done in OpenRefine or Excel, but decided to use python for consistency using a single source
def copy_year(row):
    try:
        return int(row['Type'])
    except:
        return

# apply function to copy year values in Type column to Year column, basically skipping every row that
# cannot be converted to int
df_fruitveg['Year'] = df_fruitveg.apply(lambda row: copy_year(row), axis=1)

# forward fill empty values in Year column to provide a year value for each row
df_fruitveg['Year'].ffill(inplace=True)

# make sure each value in Year column has type int
df_fruitveg['Year'] = df_fruitveg['Year'].astype(int)

# check
df_fruitveg.head()
```

```python
# note: there are some apple varieties (and other fruit) in the orchfruit dataset that is not 
# present in the other dataset, decided to only select the apple varieties that is present in both 
# and then drop all rows with missing values

# select required columns (i.e. year and apple varieties present in both datasets)
df_orchfruit = df_orchfruit[['year', 'Discovery', 'WorcesterPearmain',
                             'Cox(andclones)', 'EgremontRusset', 'Gala(andclones)',
                             'Jonagold(andclones)', 'Bramleys Seedling']]

# drop all rows with missing values
df_orchfruit.dropna(inplace=True)

# check
df_orchfruit.head()
```

#### 3. Identify and handle inconsistent or dirty data

```python
# note: there are some apple varieties in the fruitveg dataset that have inconsistent names,
# decided to rename the 'bad' names, i.e. with special characters and ending with 'group'

# rename some of the apple varieties
df_fruitveg.replace({'Apple variety' : { 'Cox’s Orange-group' : 'Cox’s Orange', 'Jonogold – group' : 'Jonogold'}}, inplace=True)

# check
df_fruitveg.head()
```

```python
# note: there are many apple varieties (and other columns) that have different names in each dataset,
# decided to rename columns in the orchfruit dataset to match the fruitveg dataset

# rename year column and some of the apple varieties (i.e. columns)
df_orchfruit.rename(columns={'year': 'Year', 'WorcesterPearmain': 'Worcester Pearmain',
                                             'Cox(andclones)': 'Cox’s Orange',
                                             'EgremontRusset': 'Egremont Russet',
                                             'Gala(andclones)': 'Gala',
                                             'Jonagold(andclones)': 'Jonogold',
                                             'Bramleys Seedling': 'Bramley’s Seedling'}, inplace=True)

# transpose dataframe to get columns as rows
df_orchfruit = df_orchfruit.transpose()

# check
df_orchfruit.head()
```

```python
# note: as noted above (from the previous check), the column names in the orchfruit dataset
# should be replaced with the values in the first row for clarity

# replace column names with values in first row
df_orchfruit.columns = df_orchfruit.iloc[0,:]

# remove first row, or select everything else
df_orchfruit = df_orchfruit[1:]

# check
df_orchfruit.head()
```

#### 4. Identify and handle ambiguity and vagueness

```python
# note: to properly analyse and plot the fruitveg dataset, decided to select the apple varieties
# which is required, then calculate the average of the average monthly prices per apple variety into 
# a new column, hence getting an average price per year an apple variety, noting that there are several 
# missing values for average monthly prices

# note: there are two types of apples in the fruitveg dataset (Cooking Apples and Dessert Apples)

# create new dataframe with dessert apples only (as required)
df_apples = df_fruitveg[df_fruitveg['Type'] == 'Dessert Apples']

# calculate average price into a new column, making sure each value has type float
df_apples.loc[:,('Average wholesale price')] = df_apples[['Jan', 'Feb', 'Mar', 'Apr',
                                                          'May', 'Jun', 'Jul', 'Aug',
                                                          'Sep', 'Oct', 'Nov', 'Dec']].astype(float).mean(numeric_only=True, axis=1)

# select required columns (and change column order) for clarity
df_apples = df_apples[['Year', 'Apple variety', 'Average wholesale price']]

# drop all rows with missing values at this point
df_apples.dropna(inplace=True)

# note: at this point, there are multiple rows for the same apple variety but different quality.

# check
df_apples.head()
```

```python
# note: there are some apple varieties in the apples dataframe that are ambiguous, decided to
# remove the apple varieties 'Other Early Season', 'Other Mid Season', and 'Other Late Season'

# select named apples only (i.e. remove others)
df_apples = df_apples[(df_apples['Apple variety'] != 'Other Early Season') &
                      (df_apples['Apple variety'] != 'Other Mid Season') &
                      (df_apples['Apple variety'] != 'Other Late Season')]

# check
df_apples.head()
```

#### 5. Put the data into an appropriate form for plotting

```python
# note: there is currently an average price per apple variety and quality in the 
# apples dataframe, decided to combine the average prices for each apple quality

# create a dataframe for grouped values, group on year and apple variety and get average prices
df_apples_grouped = df_apples.groupby(['Year', 'Apple variety']).agg('mean')

# use stack to get into a compatible format (for merge) and reset index
df_apples_stacked = df_apples_grouped.stack().reset_index()

# check
#df_apples_stacked

# rename column for clarity
df_apples_stacked = df_apples_stacked.rename(columns={0: 'Average wholesale price'})

# select required columns (and change column order) for clarity
df_apples_stacked = df_apples_stacked[['Year', 'Apple variety', 'Average wholesale price']]

# check
df_apples_stacked.head()
```

```python
# note: the year values are currently the column names in the orchfruit datafrane, decided
# to stack and transform into a compatible format (as the apples stacked dataframe)

# use stack to get into compatible format and reset index
df_orchfruit_stacked = df_orchfruit.stack().reset_index()

# check
#df_orchfruit_stacked

# rename columns into compatible format
df_orchfruit_stacked.rename(columns={'level_0': 'Apple variety', 0: 'Orchard space'}, inplace=True)

# select required columns (and change column order)
df_orchfruit_stacked = df_orchfruit_stacked[['Year', 'Apple variety', 'Orchard space']]

# sort by year
df_orchfruit_stacked.sort_values('Year', inplace=True)

# note: this fixed an odd error while trying to merge
df_orchfruit_stacked['Year'] = df_orchfruit_stacked['Year'].astype(int)

# check
df_orchfruit_stacked.head()
```

```python
# note: the apples stacked and orchfruit stacked dataframe should be fairly similar at this
# point, decided to merge on Year and Apple variety

# merge the two dataframes on Year and Apple variety
df_merged = pd.merge(df_apples_stacked, df_orchfruit_stacked, on=['Year', 'Apple variety'], how='left')

# note: include some NaNs to avoid limiting average prices analysis.

# for testing only: drop all rows with missing values at this point
#df_merged.dropna(inplace=True)

# check
df_merged.head()
```

#### 6. Visualise the data

```python
# note: this first plot is showing the average prices per apple variety and year

# note: Set year values to string to avoid odd error when combining plots.
df_merged['Year'] = df_merged['Year'].astype(str)

# group by year and apple variety for average wholesale price
df_grouped_average = df_merged[['Year', 'Apple variety', 'Average wholesale price']].groupby(['Year', 'Apple variety']).mean()

# note: values for some apple varieties are missing at some years (e.g. katy and braeburn).

# unstack and plot as multiple lines and yticks between 0.3 and 0.8
df_grouped_average['Average wholesale price'].unstack().plot.line(title="Average wholesale prices per apple variety and year (2004-2015)",
                                                                  figsize=(12,6), 
                                                                  yticks=np.arange(0.3, 0.8, 0.05))

# set x any y-labels, and position legend
plt.xlabel('Year')
plt.ylabel('Average price in GBP/kg')
plt.legend(loc='upper left')
```

```python
# note: this second plot is showing the orchard spaces per apple variety and year

# note: filter orchard space by year (or use dropna) to get a better individual plot from merged dataframe

# group by year and apple variety for orchard space
df_grouped_orchard = df_merged[['Year', 'Apple variety', 'Orchard space']].groupby(['Year', 'Apple variety']).sum()

# unstack and plot as multiple bars for years data is available
df_grouped_orchard['Orchard space'].dropna().unstack().plot.bar(title="Orchard space per apple variety and year (2004, 2007, 2009, and 2012)", figsize=(12,6))

# set x any y-labels, and position legend
plt.xlabel('Year')
plt.ylabel('Orchard space')
plt.legend(loc='upper right')
```

```python
# note: this third plot is showing average prices per apple variety and year and the corresponding
# orchard space (if applicable), decided to include all apples in this plot (somewhat messy)

# create an empty figure for combined plot
fig = plt.figure()

# axes and plot for average price, as a solid line and yticks between 0.3 and 0.8
average_axes = fig.add_subplot(111)
df_grouped_average['Average wholesale price'].unstack().plot.line(ax=average_axes,
                                                                  figsize=(15,10), 
                                                                  yticks=np.arange(0.3, 0.8, 0.05))

# axes and plot for orchard space, as multiple bars
orchard_axes = average_axes.twinx()
df_grouped_orchard['Orchard space'].unstack().plot.bar(ax=orchard_axes, figsize=(15,10), alpha=0.5)

# labels and title for combined plot
average_axes.set_ylabel('Average price in GBP/kg')
orchard_axes.set_ylabel('Orchard space')

# legend prositions for combined plot
average_axes.legend(loc=4, frameon=True, fontsize=10).set_visible(True)
orchard_axes.legend(loc=4, frameon=True, fontsize=10).set_visible(False)

# add a title
plt.title('Average prices and orchard spaces\n per apple variety and year.', fontsize=20)
```

```python
# note: since the third plot, as seen above, is quite messy, decide to create an alternative plot
# with only selected apple varieties (i.e. those present in both dataframes)

# copy merged dataframe
df_merged_alt = df_merged.copy()

# select apple varieties to include
df_merged_alt = df_merged_alt[(df_merged_alt['Apple variety'] == 'Cox’s Orange') |
                              (df_merged_alt['Apple variety'] == 'Discovery') |
                              (df_merged_alt['Apple variety'] == 'Egremont Russet') |
                              (df_merged_alt['Apple variety'] == 'Gala') |
                              (df_merged_alt['Apple variety'] == 'Jonogold') |
                              (df_merged_alt['Apple variety'] == 'Worchester Pearmain')]

# check
#df_merged_alt.head()

# group by year and apple variety for average wholesale price
df_grouped_average_alt = df_merged_alt[['Year', 'Apple variety', 'Average wholesale price']].groupby(['Year', 'Apple variety']).mean()

# group by year and apple variety for orchard space
df_grouped_orchard_alt = df_merged_alt[['Year', 'Apple variety', 'Orchard space']].groupby(['Year', 'Apple variety']).sum()
```

```python
# create an empty figure for alternative combined plot
fig = plt.figure()

# axes and plot for average price, as a solid line and yticks between 0.3 and 0.8
average_axes = fig.add_subplot(111)
df_grouped_average_alt['Average wholesale price'].unstack().plot.line(ax=average_axes,
                                                                      figsize=(15,10), 
                                                                      yticks=np.arange(0.3, 0.8, 0.05))

# axes and plot for orchard space, as multiple bars
orchard_axes = average_axes.twinx()
df_grouped_orchard_alt['Orchard space'].unstack().plot.bar(ax=orchard_axes, figsize=(15,10), alpha=0.5)

# labels and title for alternative combined plot*
average_axes.set_ylabel('Average price in GBP/kg')
orchard_axes.set_ylabel('Orchard space')

# legend prositions for alternative combined plot
average_axes.legend(loc=2, frameon=True, fontsize=10).set_visible(False)
orchard_axes.legend(loc=4, frameon=True, fontsize=10).set_visible(True)

# add a title for alternative plot
plt.title('Alternative plot:\nAverage prices and orchard spaces\n per apple variety and year.', fontsize=20)
```

#### 7. Interpret your plot

Average wholesale prices for some apple varieties are missing at certain years in the dataframe and the first combined plot, e.g. data for 'Katy' is only available for year 2004 (and hence does not appear in the plot at all) and 'Braeburn' is only available for year 2014 and 2015 (and hence only visible at the far right side in the plot), and some apple varieties are missing orchard space data, hence bars for those apple varieties are missing from the plot as well. 

In the alternative and final plot, continuing on the mentioned 'issues', selecting only the apple varieties present in both of the datasets (i.e. apple varieties with average wholesale price and orchard space).

As seen in individual and combined plots above, average prices (lines) seem to increase overall for most apple varieties (with some fluctuation) and the orchard spaces (bars) seem to be stable (or slightly decreasing), with the exception for 'Gala', which is increasing, and 'Cox's Orange', which is decreasing.
